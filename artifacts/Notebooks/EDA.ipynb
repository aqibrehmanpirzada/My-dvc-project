#### Exploratory Data Analysis (EDA) is an important step in understanding and summarizing the main characteristics of your dataset. Given your dataset with various columns, here are the steps you can take for EDA:

1. Data Cleaning:

Check for missing values in each column and decide how to handle them. You might need to impute missing values or remove rows/columns with too many missing values.
Convert data types to appropriate formats. For example, convert object types to numerical formats where applicable.
Remove any duplicate rows if present.

2. Summary Statistics:

Calculate summary statistics such as mean, median, standard deviation, minimum, and maximum for numerical columns.
For categorical columns, find unique values and their frequencies.

3. Data Visualization:

Create visualizations to understand the distribution of numerical variables using histograms, box plots, and density plots.
For categorical variables, create bar plots to visualize the frequency of each category.
Plot correlations between numerical variables using scatter plots or correlation matrices.
Geographical data like latitude and longitude can be visualized on a map.

4. Handling Outliers:

Identify and handle outliers in your numerical data. Outliers can significantly impact statistical analysis and modeling.
5. Feature Relationships:

Explore how different numerical and categorical variables are related to each other.
Check for any patterns or trends that might reveal interesting insights.
6. Data Distributions:

Analyze the distributions of variables. Check if they follow normal distributions or if they have skewed distributions.
7. Handling Categorical Variables:

Explore the distribution of categories within categorical variables.
Consider using bar plots or pie charts to visualize categorical distributions.

8. Data Transformation:

If necessary, transform variables to make their distributions more normal. Common transformations include logarithmic or power transformations.

9. Correlation Analysis:

Calculate and visualize correlations between numerical variables using correlation matrices or heatmaps.
Identify strong correlations that might indicate relationships between variables.

9. Domain-Specific Analysis:

If you have domain-specific knowledge, use it to guide your exploration. For example, if your dataset is related to economics, you might focus on variables like GDP, tax rates, and unemployment.

10. Missing Data Analysis:

Analyze the patterns of missing data. Are the missing values random or do they follow a pattern? This can provide insights into data collection processes.

11. Statistical Tests (Optional):

Perform statistical tests to assess hypotheses or relationships between variables, if applicable.
Remember that EDA is a flexible process, and the steps you take can vary based on the nature of your dataset and the insights you're seeking to gain. Visualization tools like matplotlib, seaborn, and pandas in Python can be incredibly helpful for conducting these analyses.

```python

def separate_categorical_numerical(data_frame):
    categorical_columns = []
    numerical_columns = []

    for column in data_frame.columns:
        if pd.api.types.is_categorical_dtype(data_frame[column]):
            categorical_columns.append(column)
        elif pd.api.types.is_numeric_dtype(data_frame[column]):
            numerical_columns.append(column)

    return categorical_columns, numerical_columns

```

```python
import pandas as pd

def separate_object_numerical(data_frame):
    object_columns = []
    numerical_columns = []

    for column in data_frame.columns:
        if data_frame[column].dtype == 'object':
            object_columns.append(column)
        elif pd.api.types.is_numeric_dtype(data_frame[column]):
            numerical_columns.append(column)

    return object_columns, numerical_columns

```

```python

def clean_percentage_column(column):
    cleaned_values = []

    for value in column:
        if isinstance(value, str):
            value = value.strip('%')  # Remove percentage symbol
            if value.replace('.', '', 1).isdigit():  # Check if value is numeric after removing first period
                cleaned_values.append(float(value))
            else:
                cleaned_values.append(np.nan)
        else:
            cleaned_values.append(value)

    return cleaned_values

# Apply the function to each percentage column separately
data['Agricultural Land( %)'] = clean_percentage_column(data['Agricultural Land( %)'])
data['CPI Change (%)'] = clean_percentage_column(data['CPI Change (%)'])
data['Minimum wage'] = clean_percentage_column(data['Minimum wage'])
data['Out of pocket health expenditure'] = clean_percentage_column(data['Out of pocket health expenditure'])
data['Population: Labor force participation (%)'] = clean_percentage_column(data['Population: Labor force participation (%)'])
data['Tax revenue (%)'] = clean_percentage_column(data['Tax revenue (%)'])
data['Total tax rate'] = clean_percentage_column(data['Total tax rate'])
data['Unemployment rate'] = clean_percentage_column(data['Unemployment rate'])
data['Forested Area (%)'] = clean_percentage_column(data['Forested Area (%)'])
data['Gross primary education enrollment (%)'] = clean_percentage_column(data['Gross primary education enrollment (%)'])
data['Gross tertiary education enrollment (%)'] = clean_percentage_column(data['Gross tertiary education enrollment (%)'])



```

```python
import numpy as np

def clean_comma_numeric_column(column):
    cleaned_values = []

    for value in column:
        if isinstance(value, str):
            cleaned_value = value.replace(',', '')  # Remove commas
            if cleaned_value.replace('.', '', 1).isdigit():  # Check if value is numeric after removing first period
                cleaned_values.append(int(cleaned_value))
            else:
                cleaned_values.append(np.nan)
        else:
            cleaned_values.append(value)

    return cleaned_values

# Apply the function to each column separately
data['Land Area(Km2)'] = clean_comma_numeric_column(data['Land Area(Km2)'])
data['Co2-Emissions'] = clean_comma_numeric_column(data['Co2-Emissions'])
data['Urban_population'] = clean_comma_numeric_column(data['Urban_population'])
```

```python
from sklearn.preprocessing import LabelEncoder

def encode_categorical_features(data):
    le = LabelEncoder()
    categorical_columns = data.select_dtypes(include=['object']).columns
    
    for col in categorical_columns:
        data[col] = le.fit_transform(data[col])
    
    return data
```
